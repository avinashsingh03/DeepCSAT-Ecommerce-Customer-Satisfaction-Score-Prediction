### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Handling Time vs CSAT Score (Correlation Test) Hypothesis**

Null Hypothesis (H₀)
Handling time has no relationship with CSAT score.

Alternative Hypothesis (H₁)
Handling time affects CSAT score.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

corr, p_value = pearsonr(data["connected_handling_time"], data["CSAT Score"])

print("Correlation:", corr)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Correlation Test


##### Interpretation

p < 0.05 → significant relationship

p ≥ 0.05 → no strong relationship

##### Business Impact

If handling time negatively affects CSAT, the company should reduce issue resolution time.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀
All shifts produce same CSAT scores.

H₁
At least one shift produces different CSAT scores.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

morning = data[data["Agent Shift"]=="Morning"]["CSAT Score"]
evening = data[data["Agent Shift"]=="Evening"]["CSAT Score"]
night = data[data["Agent Shift"]=="Night"]["CSAT Score"]

f_stat, p_value = f_oneway(morning, evening, night)

print("F-statistic:", f_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

ANOVA Test.

##### Interpretation

If p < 0.05, shifts perform differently.

##### Business Impact

Some shifts may need better training or staffing.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀
Agent experience does not affect CSAT score.

H₁
Agent experience affects CSAT score.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
groups = data.groupby("Tenure Bucket")["CSAT Score"].apply(list)

from scipy.stats import f_oneway

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

ANOVA Test.

##### Interpretation

If p < 0.05, experienced agents produce better satisfaction scores.

### 1. Handling Missing Values

````python


df = pd.read_csv("/content/DeepCSAT-Ecommerce-Customer-Satisfaction-Score-Prediction/cleaned_customer_support_data.csv")

df.head()
````

````python
# Handling Missing Values & Missing Value Imputation
df.isnull().sum()
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Customer Remarks columns with datatype object is filled with value "Unknown".

### 2. Convert Date Column to DateTime

````python
# Convert Date Columns to Datetime
date_cols = [
    "order_date_time",
    "Issue_reported at",
    "issue_responded",
    "Survey_response_Date"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
````

````python
df[date_cols].head()
````

### 3. Create Time-Based Features

````python
# Create Time-Based Features
df["response_time_minutes"] = (
    (df["issue_responded"] - df["Issue_reported at"]).dt.total_seconds() / 60
)

df["order_to_issue_hours"] = (
    (df["Issue_reported at"] - df["order_date_time"]).dt.total_seconds() / 3600
)

df["survey_delay_hours"] = (
    (df["Survey_response_Date"] - df["issue_responded"]).dt.total_seconds() / 3600
)
````

#### Why these features are created?

These features are very powerful predictors of customer satisfaction.
These represent:

* response_time_minutes → how quickly support responded

* order_to_issue_hours → time before issue appeared

* survey_delay_hours → time before feedback given

### 4. Extract Date Components

````python
#Extract Date Components
df["order_day"] = df["order_date_time"].dt.day
df["order_month"] = df["order_date_time"].dt.month
df["order_hour"] = df["order_date_time"].dt.hour

df["issue_day"] = df["Issue_reported at"].dt.day
df["issue_hour"] = df["Issue_reported at"].dt.hour
````

These features help detect temporal patterns.

Example:

* more complaints at night

* weekend issues

### 5. Create Price Buckets

````python
# 5. Create Price Buckets
df["price_bucket"] = pd.cut(
    df["Item_price"],
    bins=[0,1000,5000,10000,50000],
    labels=["Low","Medium","High","Premium"]
)
````

Customers buying expensive products behave differently.

### 6. Text Feature Engineering

````python
# replacing unknown with empty remarks
df["Customer Remarks"] = df["Customer Remarks"].replace("unknown", "")
df["remark_length"] = df["Customer Remarks"].astype(str).apply(len)
````

````python
# drop original column
df = df.drop(columns=["Customer Remarks"])
````

````python
df.head()
````

##### Why useful?



Long complaints often indicate high dissatisfaction.

Handling extreme values improves model stability.

### 8.Drop Unnecessary Columns

````python
df = df.drop(columns=[
    "order_date_time",
    "Issue_reported at",
    "issue_responded",
    "Survey_response_Date"
])
````

After feature creation, original datetime columns may not be needed.

## ***7. ML Model Implementation***

### ML Model - 1

````python
# ML Model - 1 Implementation

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Fit the Algorithm
lr_model = LogisticRegression(max_iter=1000)

lr_model.fit(X_train, y_train)

# Predict on the model
y_pred_lr = lr_model.predict(X_test)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# Evaluation Metrics

lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_precision = precision_score(y_test, y_pred_lr, average='weighted')
lr_recall = recall_score(y_test, y_pred_lr, average='weighted')
lr_f1 = f1_score(y_test, y_pred_lr, average='weighted')

import pandas as pd

lr_scores = pd.DataFrame({
    'Metric':['Accuracy','Precision','Recall','F1 Score'],
    'Score':[lr_accuracy, lr_precision, lr_recall, lr_f1]
})

lr_scores
````

````python
# Visualizing evaluation Metric Score chart
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Metric', y='Score', data=lr_scores)
plt.title("Logistic Regression Evaluation Metrics")
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C':[0.01,0.1,1,10]
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)

# Fit the Algorithm
grid_lr.fit(X_train, y_train)

best_lr = grid_lr.best_estimator_

# Predict on the model
y_pred_lr_tuned = best_lr.predict(X_test)
````

##### Which hyperparameter optimization technique have you used and why?

GridSearchCV was used for hyperparameter tuning. It systematically tests multiple combinations of parameters and selects the best one using cross-validation. This ensures that the model finds the optimal parameter configuration and reduces the risk of overfitting.

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

For evaluating the performance of the machine learning models, the following evaluation metrics were considered:

Accuracy, Precision, Recall, and F1 Score.

* Accuracy measures the overall percentage of correct predictions made by the model. Since the objective of this project is to correctly predict customer satisfaction scores, accuracy provides a general understanding of model performance.

* Precision indicates how many of the predicted satisfaction scores were actually correct. In a business context, higher precision ensures that the model does not incorrectly classify dissatisfied customers as satisfied, which could lead to ignoring customers who need support.

* Recall measures how well the model identifies all relevant cases of customer satisfaction levels. A higher recall ensures that the system captures most customer dissatisfaction cases so that the company can take corrective actions.

* F1 Score is the harmonic mean of precision and recall. It provides a balanced evaluation when both false positives and false negatives are important. In customer satisfaction prediction, maintaining a balance between identifying dissatisfied customers and avoiding incorrect predictions is critical for business decision making.

Using these evaluation metrics helps the company make data-driven improvements in customer support operations, leading to improved customer satisfaction and service quality.
